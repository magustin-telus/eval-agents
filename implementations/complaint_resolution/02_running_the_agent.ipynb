# 02 · Running the Complaint-Resolution Agent

The agent reads one complaint, classifies it, calls `retrieve_policy`, and returns a
**structured** result: `predicted_category`, `cited_policy_ids`, `resolution`, `reasoning`.

Requires a valid Gemini key in `.env` (ADK reads `GOOGLE_API_KEY`/`GEMINI_API_KEY`; the
retrieval path reads `OPENAI_API_KEY`).

```python
from dotenv import load_dotenv


load_dotenv()

from aieng.agent_evals.complaint_resolution.task import ComplaintResolutionTask


task = ComplaintResolutionTask()
```

## Run the agent on one complaint

```python
complaint = "charged twice same purchase credit card statement duplicate transaction disputed refund unauthorized"
result = await task(item={"input": complaint})
print(result["predicted_category"], "->", result["cited_policy_ids"])
```

## Inspect the structured output

```python
print("predicted_category:", result["predicted_category"])
print("cited_policy_ids  :", result["cited_policy_ids"])
print()
print("resolution:\n", result["resolution"])
print()
print("reasoning:\n", result["reasoning"])
```

## What happened

1. The agent classified the complaint into one of the five categories.
2. It called the `retrieve_policy` tool, which returned the matching policy document.
3. It wrote a resolution **grounded in that policy text**.

Each run is traced to Langfuse (if configured), where you can inspect the tool call and the
retrieved policy — the same trace the groundedness evaluator reads in notebook 03.

```python
await task.close()
```
